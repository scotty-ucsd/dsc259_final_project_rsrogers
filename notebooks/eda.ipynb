### Read `outage_clean.csv`

````python
data = "../data/outage_clean.csv"
df = pd.read_csv(data)
# Output dataset info, summary statistics, and null counts
print("Dataset Info:")
df.info()

print("\nMissing Values per Column:")
print(df.isna().sum()[df.isna().sum() > 0])

print("\nSummary Statistics for Numeric Columns:")
display(df.describe())


# --- 1. PRELIMINARY CLEANING ---

# A. Drop the unnecessary index column
if 'OBS' in df.columns:
    df = df.drop(columns=['OBS'])

# B. Consolidate Temporal Features into Datetime objects
df['OUTAGE.START'] = pd.to_datetime(df['OUTAGE.START.DATE'] + ' ' + df['OUTAGE.START.TIME'], errors='coerce')
df['OUTAGE.RESTORATION'] = pd.to_datetime(df['OUTAGE.RESTORATION.DATE'] + ' ' + df['OUTAGE.RESTORATION.TIME'], errors='coerce')

# Drop the redundant temporal columns
temporal_cols_to_drop = [
    'YEAR', 'MONTH', 
    'OUTAGE.START.DATE', 'OUTAGE.START.TIME', 
    'OUTAGE.RESTORATION.DATE', 'OUTAGE.RESTORATION.TIME'
]
df = df.drop(columns=[col for col in temporal_cols_to_drop if col in df.columns])

# C. Convert HURRICANE.NAMES to a boolean flag and drop the original
df['IS_HURRICANE'] = df['HURRICANE.NAMES'].notna()
df = df.drop(columns=['HURRICANE.NAMES'])

# D. Standardize categorical text (lowercase and strip whitespace)
string_cols = df.select_dtypes(include=['object']).columns
for col in string_cols:
    df[col] = df[col].str.lower().str.strip()


# --- 2. INSPECTING MISSING OUTAGE DURATIONS ---

# Isolate the rows where OUTAGE.DURATION is missing
df_missing_duration = df[df['OUTAGE.DURATION'].isna()]

print(f"Total rows missing OUTAGE.DURATION: {len(df_missing_duration)}")

# Check if these missing durations are clustered around specific causes
print("\nMissing Durations by Cause Category:")
display(df_missing_duration['CAUSE.CATEGORY'].value_counts(dropna=False))

# Check if they are clustered around specific years
print("\nMissing Durations by Start Year:")
display(df_missing_duration['OUTAGE.START'].dt.year.value_counts(dropna=False))

# Check if restoration times are also missing for these rows (indicating ongoing outages)
missing_restoration = df_missing_duration['OUTAGE.RESTORATION'].isna().sum()
print(f"\nOut of {len(df_missing_duration)} missing durations, {missing_restoration} are also missing RESTORATION times.")
````
